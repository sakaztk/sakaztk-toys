# 初期設定
* wsl --install ubuntu でユーザー名パスワードの設定
* ubuntu内でsudo設定だけする
```` 
echo 'sakaztk ALL=(ALL) NOPASSWD:ALL' | sudo tee /etc/sudoers.d/sakaztk
sudo chmod 440 /etc/sudoers.d/sakaztk
````
* 以降をホストWindowsから実行

`````powershell
wsl sudo apt-get -o Acquire::ForceIPv4=true update
`````

`````powershell
wsl sudo apt-get -o Acquire::ForceIPv4=true upgrade -y
`````

`````powershell
wsl sudo apt-get -o Acquire::ForceIPv4=true install -y python3-pip
`````

`````powershell
wsl sudo apt-get install -y pipx
`````

`````powershell
wsl pipx install pipx
`````

`````powershell
wsl sudo apt-get remove -y pipx
`````

`````powershell
wsl sudo ~/.local/bin/pipx install pipx --global
`````

`````powershell
wsl pipx uninstall pipx
`````

`````powershell
wsl pipx --version
`````
