```python
import ansible_runner
```

```python
rc = ansible_runner.runner_config.RunnerConfig(
    private_data_dir='../../',
    playbook='playbooks/setup-wsl.yml',
    inventory='inventory/hosts.yml',
    artifact_dir='./artifacts',
    verbosity=0,
    envvars={
        'ANSIBLE_CONFIG': './ansible.cfg',
        'ANSIBLE_HOST_KEY_CHECKING': 'False',
        'ANSIBLE_ROLES_PATH': '../roles:./roles',
    }
)
#rc.cmdline = '--check'
rc.prepare()
```

```python
runner = ansible_runner.Runner(config=rc)
result = runner.run()
```

```python
print("UUID:", rc.ident)
print("Artifacts Dir:", rc.artifact_dir)
print("Status:", runner.status)
print("RC:", runner.rc)
```
